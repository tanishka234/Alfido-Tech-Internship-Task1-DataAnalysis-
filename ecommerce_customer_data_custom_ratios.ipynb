# üìä Task 1(a):Customer Behviour Analysis

## üéØ Objective
Analyze customer behavior using an e-commerce dataset to uncover:
- Who the customers are (gender, age)
- What they buy (categories)
- How much they spend
- Who is more likely to churn


## üõ†Ô∏è Technologies Used
- Python (Pandas)
- Matplotlib,seaborn
- Google Colab

```python
import pandas as pd

df = pd.read_csv("ecommerce_customer_data_custom_ratios.csv")
df.head()
```

```python
print("Missing values:\n", df.isnull().sum())
df = df.dropna()

print("Duplicates:", df.duplicated().sum())
df = df.drop_duplicates()

df['Purchase Date'] = pd.to_datetime(df['Purchase Date'])

df['Total Purchase Amount'] = pd.to_numeric(df['Total Purchase Amount'], errors='coerce')
```

```python
print("Gender distribution:\n", df['Gender'].value_counts())
print("Age stats:\n", df['Age'].describe())
print("Top Product Categories:\n", df['Product Category'].value_counts())
print("Churn Rate:", df['Churn'].mean() * 100, "%")
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.countplot(x='Gender', data=df)
plt.title("Gender Distribution")
plt.show()


sns.histplot(df['Age'], bins=15, kde=True)
plt.title("Age Distribution")
plt.show()

sns.countplot(y='Product Category', data=df, order=df['Product Category'].value_counts().index)
plt.title("Top Product Categories")
plt.show()

sns.boxplot(x='Gender', y='Total Purchase Amount', data=df)
plt.title("Purchase Amount by Gender")
plt.show()

plt.figure(figsize=(10,6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```

# üìä Customer Behavior Analysis ‚Äì Summary Report

## üîπ Gender Distribution
- The customer base is **almost evenly split** between male and female customers.
- Both genders exhibit **similar purchase patterns**, suggesting gender-neutral marketing strategies may be effective.

---

## üîπ Age Distribution
- Customers span a wide age range, primarily from **18 to 70 years old**.
- The distribution is **fairly uniform**, indicating that users of all ages are engaging with the platform.
- This implies a need for **diverse product offerings** that appeal to different age groups.

---

## üîπ Top Product Categories
- The most purchased categories are:
  1. **Clothing**
  2. **Books**
  3. **Electronics**
  4. **Home**
- Clothing and books **dominate** the product choices, indicating customer preference toward lifestyle and leisure items.

---

## üîπ Purchase Amount by Gender
- **Median purchase values** are nearly equal for male and female customers.
- The **range of purchase amounts** is also similar, suggesting both genders contribute equally to revenue.
---


## üîπ Correlation Heatmap Insights
- No strong correlation was observed between variables.
- A **slight positive correlation** exists between `Customer Age` and `Total Purchase Amount` (0.052), implying older customers might spend slightly more.
- `Returns` and `Churn` show **negligible correlations**, indicating independent behavior.

---

## ‚úÖ Final Insights & Recommendations

- üìå **Target marketing can remain gender-neutral** due to balanced participation.
- üéØ Focus on enhancing product variety in **Clothing** and **Books**, the most popular categories.
- üìà Consider **age-based personalization** for promotions to leverage the slightly higher spending trend among older customers.
- üí° No urgent indicators of churn or dissatisfaction, but periodic analysis is advised.
